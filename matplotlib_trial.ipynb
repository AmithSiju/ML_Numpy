4. Subplots

```python
a = np.arange(0,5,0.1)
b = np.sin(a)
c = np.cos(a)
plt.subplot(2,1,1)
plt.plot(a,b)
plt.title("Sin Wave")
plt.subplot(2,1,1)
plt.plot(a,c)
plt.title("Cos Wave")

plt.subplot(2,1,1)
plt.plot(b,c)
plt.title("Wave")
# plt.subplot(a,c)
# plt.subplot(b,c)
```

Q. Write a python code to plot a line chart to debit the run-rate of T20 Match from the given Data. 
    Overs | Runs
    05    | 045 
    10    | 079 
    15    | 145
    20    | 234

Also change the thickness of the line.

```python
x = [5,10,15,20]
y = [45,79,145,234]
plt.title("Run-Rate")
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.plot(x,y,'--c',linewidth=2)
```
