Linear Regression

```python
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X=2*np.random.rand(100,1)
y=4+3*X+np.random.rand(100,1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
m= LinearRegression()
m.fit(X_train,y_train)
y_pred=m.predict(X_test)
r=r2_score(y_pred,y_test)
print(r)
z=mean_squared_error(y_pred,y_test)
print(z)
```

```python
plt.scatter(X,y, color="red" ,marker="*", label="actual")
plt.plot(X_test,y_pred, color="blue", label="prediction")
plt.legend()
plt.show()
```

```python
n=float(input("Enter a value to predict : "))
n1=np.array([[n]])
result=m.predict(n1)
print(result)
```

K-means

```python
import numpy as np
from sklearn.datasets import make_blobs #used to produce sample data 
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
import matplotlib.pyplot as plt
```

```python
data,_ = make_blobs(n_samples=300)
m=KMeans(n_clusters=5,random_state=42,n_init="auto") #auto used for initializing centroid
m.fit(data)
labels=m.labels_ # contains cluster assignment for each data point .
set(labels) # set the labels for each data points.

c = m.cluster_centers_ #stores the co-ordinates of the centroids for these 5 clusters.
plt.scatter(data[:,0],data[:,1],c=labels,cmap="viridis", label="data") #
plt.scatter(c[:,0],c[:,1],color="red",label="centroid")
plt.legend(list(labels))
plt.show()
m.predict([[2,-4],[6,6],[6,-2]])
```
