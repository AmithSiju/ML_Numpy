SVM :
    # SVM is a supervised learning algorithm, which is used for classification purpose. 
    # It is used both for linear and non-linear classification,
    # Used for regression problem
    # Used for outlier detection task
    Application :
        -> Setiment analysis
        -> Image recognition
        -> Handwritting identification
        -> Spam detection
        -> Face detection
        -> Anomaly detedction
        -> Gene expression analyasis

SVM Linear:
    # It is a 2-D surafce, a straight line that classifies data points
    # Also called as max. margin linear classifier-setting max. margin. It ensure the margin b/w the closest datapoints of diff. classes(support vectors)     is maximized

SVM Non-Linear:
    # Instead of a straight line , we are using a hyperplan. The dimension of hyperplan depends on the no. of features.
    # There can be multiple planes or decision boundries to segregate the classes in n-dimensional space - we need to find out the best decision boundary that helps to classify the datapoints. This best boundary is called hyper plane. 
    # Simply SVM is a ML Algorithm to find the hyperplane in any n-dimensional plane that distictly classify the datapoints
    # When outliers are there need to mis-classify or ignore it.
    # If we consider outliers - min. margin - wrong prediction


Soft-Margin Formulation :
    It allows SVM to make a certain no. of mistakes and keep margin as wide as possible, so that other datapoints can still be classified correctly.

? How to perform mis-classification ?
Ans : With the help of "c-parameter". It tells the SVM how much you want to avoid mis-classifying in each section. When you increase the value of c-parameter min. margin occurs and vice versa.

Errors in SVM :
    # CLassification error (A) 
    # Margin error (B)
    SVM goal = min(cA+B)

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_samples = 1000, n_features=2, n_informative=2, n_redundant=0, random_state=42)
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
a=SVC(kernel="linear" )
a.fit(X_train,y_train)
y_pred= a.predict(X_test)
b=accuracy_score(y_test,y_pred)
print(b)
r=classification_report(y_test,y_pred)
print(r)
print(f"Classification report {r}")
```

```python
#plotting the decision boundary
plt.figure(figsize=(8,6))
```

```python
plt.scatter(X[:,0],X[:,1],cmap="viridis",s=50,alpha=0.6)
plt.gca()
```

```python
#plotting boundary
plt.scatter(X[:,0],X[:,1],c=y, cmap="viridis",s=50,alpha=0.6)
plt.gca()
ax=plt.gca()
d=ax.get_xlim()
e=ax.get_ylim()
xx,yy=np.meshgrid(np.linspace(d[0],d[1],100),np.linspace(e[0],e[1],100))
z=a.decision_function(np.c_[xx.ravel(),yy.ravel()])
z=z.reshape(xx.shape)
plt.contour(xx,yy,z,colors='k',levels=[-1,0,1],alpha=0.5,linestyles=['--','-','--'])
plt.title("SVM Decision Boundary ")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
